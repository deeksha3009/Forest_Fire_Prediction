```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import math
from math import exp
from sklearn.svm import SVR
from pandas import DataFrame, read_csv

fire = read_csv('forestfires.csv')
fire.head()
fire.describe()
```

```python
#we transform the outcome 'area' by taking its logarithm (after adding 1 to avoid zeros)
fire['Log-area']=np.log10(fire['area']+1)
for i in fire.describe().columns[:-2]:
    fire.plot.scatter(i,'Log-area')
```

```python
#splitting into train and test set
from sklearn.model_selection import train_test_split
features = fire.drop(['area','Log-area'],axis=1)
target = fire['Log-area']
x_train, x_test, y_train, y_test = train_test_split(features,target, test_size=0.2,random_state=50)
```

```python
def rec(m,n,tol):
    if type(m)!='numpy.ndarray':
        m=np.array(m)
    if type(n)!='numpy.ndarray':
        n=np.array(n)
    l=m.size
    percent = 0
    for i in range(l):
        if np.abs(10**m[i]-10**n[i])<=tol:
            percent+=1
    return 100*(percent/l)
```

```python
# Define the max tolerance limit for REC curve x-axis
tol_max=20
```

```python
from sklearn.svm import SVR

model = SVR(gamma=10)
model.fit(x_train, y_train)
print("Accuracy",model.score(x_train, y_train))
a = model.predict(x_test)
print("RMSE for Support Vector Regression:",np.sqrt(np.mean((y_test-a)**2)))
```

```python
plt.xlabel("Actual area burned")
plt.ylabel("Error")
plt.scatter(10**(y_test),10**(a)-10**(y_test))
```

```python
rec_SVR=[]
for i in range(tol_max):
    rec_SVR.append(rec(a,y_test,i))

plt.figure(figsize=(5,5))
plt.xticks([i*5 for i in range(tol_max+1)])
plt.ylim(-10,100)
plt.yticks([i*20 for i in range(6)])
plt.plot(range(tol_max),rec_SVR)
```
